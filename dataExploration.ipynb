```python
import glob
import os
import pandas as pd
import sys
```

```python
absPath = os.path.abspath(sys.path[0])
files = glob.glob(os.path.join(absPath, "*/*/*/*.csv"))

data = []
for file in files:
    df = pd.read_csv(file)
    data.append(df)

dataFrame = pd.concat(data, axis=1)
```
